## annual_supply_raw.csv

```python
import pandas as pd

pd.set_option("display.max_columns", None)

path = "data/raw/nass/annual_supply_raw.csv"
df = pd.read_csv(path, low_memory=False)
```

```python
df.info()
print("Shape:", df.shape)
display(df.head(),)
```

```python
df["Value_num"] = (
    df["Value"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(r"[^0-9.\-]", "", regex=True)
)
df["Value_num"] = pd.to_numeric(df["Value_num"], errors="coerce")
df.info()
```

```python
pd.set_option("display.max_colwidth", None)
```

```python
# 1. 每一列的类型、缺失率、唯一值数量、样例

profile = []

for col in df.columns:
    s = df[col]
    row = {
        "column": col,
        "dtype": str(s.dtype),
        "non_null": int(s.notna().sum()),
        "missing_pct": round(s.isna().mean() * 100, 2),
        "n_unique": int(s.nunique(dropna=True)),
        "sample_values": s.dropna().astype(str).unique()[:8].tolist(),
    }

    if pd.api.types.is_numeric_dtype(s):
        row.update({
            "min": s.min(),
            "median": s.median(),
            "max": s.max(),
        })
    else:
        row.update({
            "min": None,
            "median": None,
            "max": None,
        })

    profile.append(row)


profile_df = pd.DataFrame(profile)
display(profile_df)

```

```python
# 2. 重点分类列有哪些取值
key_cols = [
    "commodity_desc",
    "statisticcat_desc",
    "unit_desc",
    "short_desc",
    "agg_level_desc",
    "state_alpha",
    "reference_period_desc",
    "year",
]

for col in key_cols:
    print(f"\n## {col}")
    display(df[col].value_counts(dropna=False).head(30).to_frame("count"))
```

```python
# 3. 看不同指标的 Value 范围
value_summary = (
    df.groupby(["commodity_desc", "statisticcat_desc", "unit_desc"], dropna=False)
    .agg(
        rows=("Value_num", "size"),
        non_null=("Value_num", "count"),
        min_value=("Value_num", "min"),
        median_value=("Value_num", "median"),
        max_value=("Value_num", "max"),
    )
    .reset_index()
    .sort_values(["commodity_desc", "statisticcat_desc"])
)

display(value_summary)
```

```python
# 4. 看 short_desc 粒度下的范围，更接近业务指标
short_desc_summary = (
    df.groupby(["short_desc", "unit_desc"], dropna=False)
    .agg(
        rows=("Value_num", "size"),
        years=("year", lambda x: f"{x.min()}-{x.max()}"),
        states=("state_alpha", "nunique"),
        min_value=("Value_num", "min"),
        median_value=("Value_num", "median"),
        max_value=("Value_num", "max"),
    )
    .reset_index()
    .sort_values("short_desc")
)

display(short_desc_summary)

```

```python
# 5. 只看你 dashboard 最常用的最终年度数据
annual_final = df[
    (df["reference_period_desc"] == "YEAR")
    & (df["agg_level_desc"].isin(["NATIONAL", "STATE"]))
].copy()

pivot = annual_final.pivot_table(
    index=["commodity_desc", "year", "state_alpha", "state_name"],
    columns="statisticcat_desc",
    values="Value_num",
    aggfunc="first",
).reset_index()

display(pivot.sort_values(by=["commodity_desc", "year", "AREA HARVESTED"], ascending=[True, True, False]))
```

## crop_progress_condition_raw.csv

```python
path_crop_progress = "data/raw/nass/crop_progress_condition_raw.csv"
df_crop_progress = pd.read_csv(path_crop_progress, low_memory=False)

# Basic info
print("=== CROP PROGRESS CONDITION DATA ===")
df_crop_progress.info()
print("Shape:", df_crop_progress.shape)
display(df_crop_progress.head(10))
```

```python
# Data cleaning - convert Value column to numeric
df_crop_progress["Value_num"] = (
    df_crop_progress["Value"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(r"[^0-9.\-]", "", regex=True)
)
df_crop_progress["Value_num"] = pd.to_numeric(df_crop_progress["Value_num"], errors="coerce")
```

```python
# Profile each column
profile_crop = []
for col in df_crop_progress.columns:
    s = df_crop_progress[col]
    row = {
        "column": col,
        "dtype": str(s.dtype),
        "non_null": int(s.notna().sum()),
        "missing_pct": round(s.isna().mean() * 100, 2),
        "n_unique": int(s.nunique(dropna=True)),
        "sample_values": s.dropna().astype(str).unique()[:8].tolist(),
    }

    if pd.api.types.is_numeric_dtype(s):
        row.update({
            "min": s.min(),
            "median": s.median(),
            "max": s.max(),
        })
    else:
        row.update({
            "min": None,
            "median": None,
            "max": None,
        })

    profile_crop.append(row)

profile_crop_df = pd.DataFrame(profile_crop)
display(profile_crop_df)
```

```python
# Key categorical columns analysis
key_cols_crop = [col for col in df_crop_progress.columns 
                 if df_crop_progress[col].dtype == 'object' or 
                 df_crop_progress[col].nunique() < 50]

print("\n=== KEY CATEGORICAL VALUES ===")
for col in key_cols_crop[:10]:  # Limit to first 10
    print(f"\n## {col}")
    display(df_crop_progress[col].value_counts(dropna=False).head(20).to_frame("count"))


# Value range by key dimensions
print("\n=== VALUE RANGES BY DIMENSIONS ===")
crop_progress_summary = (
    df_crop_progress.groupby(["commodity_desc", "statisticcat_desc", "unit_desc"], dropna=False)
    .agg(
        rows=("Value_num", "size"),
        non_null=("Value_num", "count"),
        min_value=("Value_num", "min"),
        median_value=("Value_num", "median"),
        max_value=("Value_num", "max"),
    )
    .reset_index()
    .sort_values(["commodity_desc", "statisticcat_desc"])
)
display(crop_progress_summary)
```
